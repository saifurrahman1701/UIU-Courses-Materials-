```python
# Graph = {'S': [['A', 1], ['B', 5], ['C', 8]],
#         'A': [['S', 1], ['D', 3], ['E', 7], ['G', 9]],
#         'B': [['S', 5], ['G', 4]],
#         'C': [['S', 8], ['G', 5]],
#         'D': [['A', 3]],
#         'E': [['A', 7]],
#         'G': [['A', 9], ['B', 4], ['C', 5]]
# }
# Heuristics = {
#   'S': 8,
#   'A': 6,
#   'B': 3,
#   'C': 3,
#   'D': 10,
#   'E': 15,
#   'G': 0
# }

Graph = {
    'S': [['B', 4], ['C', 3]],
    'B': [['S', 4], ['F', 5], ['E', 12]],
    'C': [['S', 3], ['D', 7], ['E', 10]],
    'D': [['C', 7], ['E', 2]],
    'E': [['B', 12], ['C', 10], ['G', 5], ['D', 2]],
    'F': [['B', 5], ['G', 16]],
    'G': [['E', 5], ['F', 16]]
}

Heuristics = {
    'S': 14,
    'B': 12,
    'C': 11,
    'D': 6,
    'E': 4,
    'F': 11,
    'G': 0
}

came_from = {}

g_score = {node: float('inf') for node in Graph}
g_score['S'] = 0;

f_score = {node: float('inf') for node in Graph}
f_score['S'] = Heuristics['S']

from queue import PriorityQueue

pq = PriorityQueue()
pq.put(('S', Heuristics['S']))

while pq.empty() == False and pq.queue[0][0] != 'G':
  item = pq.get()
  for child in Graph[item[0]]:
    temp_g_score = g_score[item[0]] + child[1];
    temp_f_score = temp_g_score + Heuristics[child[0]]
    if temp_f_score < f_score[child[0]]:
      g_score[child[0]] = temp_g_score
      f_score[child[0]] = temp_f_score
      pq.put((child[0], f_score[child[0]]))
      came_from[child[0]] = item[0]

Path = []
item = 'G'
while item != 'S':
  Path.append(item)
  item = came_from[item]
Path.append('S')
print(Path[::-1])




```
